## **Cybersecurity Threats: A Decade of Digital Warfare (2015-2024)**  
Data source --- 
The global cybersecurity landscape has been shaped by relentless digital attacks over the past decade. Analyzing **3,000 recorded cyber incidents** across **10 countries**—including the **USA, China, India, UK, Germany, and Japan**—reveals critical patterns in the evolution of cyber threats.  

### **A Global Battlefield: Who is Under Attack?**  
- The dataset spans **Asia, Europe, North America, and Australia**, with **Africa notably absent**, raising concerns about underreporting or data exclusion.  
- Cyberattacks remained consistently high, fluctuating **between 290 and 300 attacks per year**, except for **2017**, which saw the highest surge (**319 attacks**), and **2019**, the lowest (**269 attacks**).  
- A pivotal moment was **2020 (COVID-19 pandemic)**, where cyber threats surged to **315 attacks**, exposing vulnerabilities in remote work environments and healthcare infrastructure.  

### **The Attack Arsenal: How Are Industries Targeted?**  
- The data identifies **six major attack types**:  
  - **Phishing** (targeting Banking, IT, and Retail most aggressively)  
  - **Ransomware**  
  - **Man-in-the-Middle**  
  - **DDoS** (Disrupting IT & Telecommunications sectors)  
  - **SQL Injection**  
  - **Malware**  
- **The Banking and IT sectors remain the prime battlegrounds**, experiencing the highest number of attacks, particularly from phishing and DDoS campaigns.  

### **The Cost of Cybercrime: Financial Fallout**  
- A **Financial Loss Heatmap** reveals that **phishing and DDoS attacks account for the highest monetary damage**, crippling industries with breaches, fraud, and infrastructure downtimes.  
- Attack sources contributing to financial devastation include:  
  - **Nation-state actors** ($26.47M average loss per attack)  
  - **Unknown sources** ($25.6M)  
  - **Insider threats** ($25.07M)  
  - **Hacker groups** ($22.87M)  

### **Mapping the Damage: Global Insights**  
A **Choropleth Map visualization** uncovers financial losses by region, illustrating the varying impact of cyber threats across different economies. The insights suggest a **direct correlation between attack sophistication and economic loss**, emphasizing the need for **advanced cybersecurity defenses and regulatory frameworks**.  

### **Final Thought: Are We Winning the Cyber War?**  
With cyberattacks **persisting at an alarming rate**, organizations must rethink their cybersecurity strategies. As AI-driven threats and state-sponsored cyber warfare continue to evolve, **threat intelligence, predictive analytics, and global cooperation** are the only ways forward.

## Split: train-test 

```python
target = "Financial Loss (in Million $)"
feature = ["Attack Type"]

y = df[target] 
X = df[feature]
```

```python
# 80:20 
spliter = int(len(X) * 0.8)
# spliter

# to train and test 
X_train, y_train = X.iloc[:spliter], y.iloc[:spliter]
X_test, y_test = X.iloc[spliter:], y.iloc[spliter:]

# X_test
```

## Baseline

```python
y_pred_baseline = [y_train.mean()] * len(y_train)
mae_baseline = mean_absolute_error(y_train, y_pred_baseline)

print("Mean Financial Loss", y_train.mean())
print("Mae Baseline", mae_baseline)
```
